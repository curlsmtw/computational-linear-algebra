# Example 1: 
Visualize the transformation of a unit sphere in $\mathbb{R}^3$ under a $(2 \times 3)$ matrix $A=\begin{bmatrix}1.50 & 0.75\\-0.50 & -1.00\end{bmatrix}$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[1.50, 0.75], [-0.50, -1.00]])

# Build the unit circle
t = np.linspace(0, 2 * np.pi, 100)
x = np.cos(t)
y = np.sin(t)
npts = len(t)

# Prepare the transformation
Ax = np.zeros(npts)
Ay = np.zeros(npts)

# Image of the unit circle under A
for i in range(npts):
    v = np.array([x[i], y[i]])
    w = np.dot(A, v)
    Ax[i] = w[0]
    Ay[i] = w[1]

# Plot the original circle and its image
plt.plot(x, y, label="Unit Circle", color='b')
plt.plot(Ax, Ay, '.', markersize=10, linewidth=3, label="Transformed Image")
plt.grid(True)
plt.axis('equal')
plt.title('Action of a linear transformation on the unit circle')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Perform Singular Value Decomposition (SVD)
U, S, Vt = np.linalg.svd(A)
print("The singular value decomposition for A is:")
print("U =\n", U)
print("S =\n", np.diag(S))
print("V =\n", Vt)

# Show the plot
plt.savefig('unit_circle_transf.png',dpi=300)
plt.show()
```

# Example 2:
Visualize the transformation of a unit sphere in $\mathbb{R}^3$ under a $(2 \times 3)$ matrix $A$

 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the matrix A
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Generate points on a unit sphere in R^3
phi = np.linspace(0, np.pi, 50)  # Azimuthal angle
theta = np.linspace(0, 2 * np.pi, 50)  # Polar angle
phi, theta = np.meshgrid(phi, theta)

# Parametric equations for the sphere
x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)

# Flatten and combine into a (3, N) array of points
X = np.vstack([x.flatten(), y.flatten(), z.flatten()])

# Apply A to map points to R^2
AX = A @ X
Ax = AX[0, :].reshape(x.shape)
Ay = AX[1, :].reshape(y.shape)

# Plot the prototype (unit sphere) in R^3
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='lightgray', alpha=0.5, edgecolor='k')

# Plot the image as a transformed ellipse in 3D
ax.plot_wireframe(Ax, Ay, np.zeros_like(Ax), color='royalblue', alpha=0.7)

# Plot settings
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')
ax.set_title('Unit Sphere and Its Image Under A')

# Set limits and aspect ratio
ax.set_xlim([-7.5, 7.5])
ax.set_ylim([-7.5, 7.5])
ax.set_zlim([-7.5, 7.5])
plt.savefig('unit_sphere_transf.png',dpi=300)
plt.show()

# Perform Singular Value Decomposition (SVD)
U, S, Vt = np.linalg.svd(A)
print("The singular value decomposition for A is:")
print("U =\n", U)
print("S =\n", np.diag(S))
print("V =\n", Vt)

```
